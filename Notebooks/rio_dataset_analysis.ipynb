```python
# add stops to exact matches
bus_data_prep_1 = bus_data_prep_0.copy()
bus_data_prep_1 = bus_data_prep_1[bus_data_prep_1['line'].isin([6 , 7 , 10])]

bus_stop_prep_1 = bus_stop_prep_0.copy()
bus_stop_prep_1 = bus_stop_prep_1[bus_stop_prep_1['line_number'].isin([6 , 7 , 10])]

bus_stop_dict = {(row['stop_latitude'], row['stop_longitude']): row['stop_name'] for _, row in bus_stop_prep_1.iterrows()}
bus_data_prep_1['stop_name'] = bus_data_prep_1.apply(lambda row: bus_stop_dict.get((row['latitude'], row['longitude']), None), axis=1)
bus_data_prep_1.isna().sum()
```

```python
# add stops to dataset
import pandas as pd
from geopy.distance import geodesic 

# function to find nearest bus stop within 30m radius
def get_nearest_stop(lat, lon, bus_stops):
    # min_dist = float("inf")
    nearest_stop = None
    
    for _, stop in bus_stops.iterrows():
        stop_coords = (stop['stop_latitude'], stop['stop_longitude'])
        gps_coords = (lat, lon)
        distance = geodesic(gps_coords, stop_coords).meters
        
        if distance <= 30:
            # min_dist = distance
            nearest_stop = stop['stop_name']
            print('done ',stop['stop_name'])
            return nearest_stop
    
    return None  

bus_data_prep_2 = bus_data_prep_1.copy()
bus_data_prep_2 = bus_data_prep_2[bus_data_prep_2['line'].isin([6 , 7 , 10])]

bus_stop_prep_2 = bus_stop_prep_1.copy()
bus_stop_prep_2 = bus_stop_prep_2[bus_stop_prep_2['line_number'].isin([6 , 7 , 10])]

# bus_data_prep_1['bus_stop_name'] = bus_data_prep_1.apply(lambda row: get_nearest_stop(row['latitude'], row['longitude'], bus_stop_prep_0), axis=1)
bus_data_prep_2['stop_name'] = bus_data_prep_2.apply(
    lambda row: row['stop_name'] if pd.notna(row['stop_name']) 
    else get_nearest_stop(row['latitude'], row['longitude'], bus_stop_prep_2),
    axis=1
)
```
