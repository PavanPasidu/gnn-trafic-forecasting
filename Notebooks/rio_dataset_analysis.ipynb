```python
# add stops to dataset
import pandas as pd
from geopy.distance import geodesic 

# function to find nearest bus stop within 30m radius
def get_nearest_stop(lat, lon, bus_stops):
    # min_dist = float("inf")
    nearest_stop = None
    
    for _, stop in bus_stops.iterrows():
        stop_coords = (stop['stop_latitude'], stop['stop_longitude'])
        gps_coords = (lat, lon)
        distance = geodesic(gps_coords, stop_coords).meters
        
        if distance <= 30:
            # min_dist = distance
            nearest_stop = stop['stop_name']
            print('done ',stop['stop_name'])
            return nearest_stop
    
    return None  

bus_data_prep_2 = bus_data_prep_1.copy()
bus_data_prep_2 = bus_data_prep_2[bus_data_prep_2['line'].isin([624 , 371 , 917])]

bus_stop_prep_2 = bus_stop_prep_1.copy()
bus_stop_prep_2 = bus_stop_prep_2[bus_stop_prep_2['line_number'].isin([624 , 371 , 917])]

# bus_data_prep_1['bus_stop_name'] = bus_data_prep_1.apply(lambda row: get_nearest_stop(row['latitude'], row['longitude'], bus_stop_prep_0), axis=1)
bus_data_prep_2['stop_name'] = bus_data_prep_2.apply(
    lambda row: row['stop_name'] if pd.notna(row['stop_name']) 
    else get_nearest_stop(row['latitude'], row['longitude'], bus_stop_prep_2),
    axis=1
)
```

```python
import numpy as np
# import cupy as cp
import pandas as pd
from scipy.spatial import cKDTree
from geopy.distance import geodesic
```

```python
# read datasets
stops_df = pd.read_csv('../bus_stops_rio_de_janeiro.csv',delimiter="|")
trajectory_df = pd.read_csv('../bus_data_prep_0.csv')
```

```python
# clean stop dataset
stops_df = stops_df[stops_df['line_number'].str.match(r'^-?\d+(\.\d+)?$', na=False)]
stops_df['line_number'] = stops_df['line_number'].astype(float)
```

```python
# logic to add rest of the stops to each trip
def process_line(group):
    stops_line = stops_df[stops_df["line_number"] == group["line"].iloc[0]]
    
    if stops_line.empty:
        group["nearest_stop_name"] = None
        group["is_at_stop"] = False
        return group

    traj_coords = np.vstack((group["latitude"], group["longitude"])).T
    stop_coords = np.vstack((stops_line["stop_latitude"], stops_line["stop_longitude"])).T

    tree = cKDTree(stop_coords)
    distances, indices = tree.query(traj_coords)

    distance_threshold = 0.0003
    group["nearest_stop_name"] = stops_line.iloc[indices]["stop_name"].values
    group["is_at_stop"] = distances <= distance_threshold
    group["distance_to_next_stop"] = distances
    return group

trajectory_df = trajectory_df.groupby("line", group_keys=False).apply(process_line)
trajectory_df.head()
```

```python
# extract day, hour, month
bus_data_prep = trajectory_df.copy()
bus_data_prep['datetime'] = pd.to_datetime(bus_data_prep['datetime'])
bus_data_prep['month'] = bus_data_prep['datetime'].dt.month
bus_data_prep['day'] = bus_data_prep['datetime'].dt.day
bus_data_prep['hour'] = bus_data_prep['datetime'].dt.hour
bus_data_prep.head()
```

```python
# add travel times
def convert_distances_to_meters(df):
    latitudes = df["latitude"].values
    lon_distances = df["distance_to_next_stop"].values
    
    lat_to_meters = 111320 
    lon_to_meters = 111320 * np.cos(np.radians(latitudes))
    
    lon_distances_meters = lon_distances * lon_to_meters
    df["distance_to_next_stop"] = lon_distances_meters
    return df

trajectory_df = convert_distances_to_meters(trajectory_df)
trajectory_df['speed'] = trajectory_df['speed'] / 3.6
trajectory_df.head()
```
